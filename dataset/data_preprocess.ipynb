```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
recent_activity = pd.read_csv('recent_activity.csv')
data = pd.read_csv('dataset.csv')
user_profile = pd.read_csv('user_Profiles.csv')
```

```python
data.shape
```

```python
data = data.drop_duplicates(subset=['Name'],keep='first')
data.shape
```

```python
user_id_list = recent_activity['User_Id'].to_list()
meal_id_list = recent_activity['Meal_Id'].to_list()
num_list = []
for x in range(0, 101):
    num_list.append(x)
```

```python
new_dataframe = pd.DataFrame(index = num_list,columns=['User_Id','Disease'])
```

```python
new_dataframe
```

```python
a = 0
for id in user_id_list:
     new_dataframe.loc[a, 'User_Id'] = id
     new_dataframe.loc[a, 'Disease'] = user_profile[user_profile['User_Id']==id]['Disease'].to_list()
     a += 1
new_dataframe
```

```python
food_dataframe = pd.DataFrame(index=num_list,columns=["food_name"])
```

```python
a = 0
for id in meal_id_list:
    food_dataframe.loc[a, 'food_name'] = data[data['Meal_Id']==id]['Name'].to_list()
    a += 1
food_dataframe
```

```python
a = food_dataframe['food_name']
new_dataframe = new_dataframe.join(a)
new_dataframe
```

```python
recent_df = recent_activity[['Rated','Liked','Searched','Purchased']]
```

```python
complete_df = new_dataframe.join(recent_df)
complete_df
```

```python
complete_df.to_csv('user_activities.csv')
```
